```python
# Look at the important learning time variables

fig, ax = plt.subplots(nrows=2, figsize = [8,8])

ax[0].hist(data=Pisa_result,x="MMINS",bins=80);
ax[0].set_xlabel('{} (mins)'.format("Math Learning Time Per Week"));

ax[1].hist(data=Pisa_result,x="OUTHOURS",bins=80);
ax[1].set_xlabel('{} (hours)'.format("Out-of-School Learning Time Per Week"));
```

On these initial math learning time and out-of-school learning plots, most of the data is set to the far left of their axes, suggesting some outliers on the right.We can also notice this on above description table.It's worth taking a bit of time to identify these outliers and see if they need to be filtered out of the data.

```python
high_outliers = ((Pisa_result["MMINS"] > 1500) | (Pisa_result["OUTHOURS"] > 60))
Outliers_df=Pisa_result.loc[high_outliers,:]
Outliers_df
```

```python
Outliers_df.COUNTRY.nunique(),Outliers_df.shape[0]
```

It doesn't make much sense to me that these students have studied Out-of-School more than 60 hours per week, it's probably mis-interpreted as minutes. More than 1500 minutes of Math Learning Time Per Week is also not representative. Since these outliers have data from all countries, I think it's better to remove these inaccurate data points.

```python
Pisa_result=Pisa_result.loc[-high_outliers,:]
```

```python
# Look at the distribution again after removing those outliers
fig, ax = plt.subplots(nrows=2, figsize = [8,8])

ax[0].hist(data=Pisa_result,x="MMINS",bins=80);
ax[0].set_xlabel('{} (mins)'.format("Math Learning Time Per Week"));

ax[1].hist(data=Pisa_result,x="OUTHOURS",bins=80);
ax[1].set_xlabel('{} (hours)'.format("Out-of-School Learning Time Per Week"));
```

After removing the high-outliers, both study time distributions look like slightly right skewed unimodals.

```python
# correlation plot

plt.figure(figsize=[10,5])
base_color = sb.color_palette()[0]

plt.subplot(1, 2, 1)
ax1=sb.boxplot(data = Pisa_result, x = 'GENDER', y = 'ANCBELONG', color = base_color);
ax1.set_ylabel("Sense of Belonging to School");

plt.subplot(1, 2, 2)
ax2=sb.violinplot(data = Pisa_result, x = 'GENDER', y = 'ANCBELONG', color = base_color);
ax2.set_ylabel("Sense of Belonging to School");
```

```python
# plot matrix: sample 2000 pisa data so that plots are clearer and they render faster

samples = np.random.choice(Pisa_result.shape[0], 2000, replace = False)
Pisa_sample = Pisa_result.loc[samples,:]

g = sb.PairGrid(data = Pisa_sample, vars = numeric_vars_anchored)
g = g.map_diag(plt.hist, bins = 30);
g.map_offdiag(plt.scatter);
```

## Multivariate Exploration

The main thing I want to explore in this part of the analysis is how the two categorical measures of gender and OECD play into the relationship between students' attitude and sense of belonging to school.

```python
# plot 5000 pisa sample data so that plots are clearer and they render faster

vars_target=["ANCSTUDREL","ANCATSCHL","ANCATTLNACT"]
samples = np.random.choice(Pisa_result.shape[0], 5000, replace = False)
Pisa_sample = Pisa_result.loc[samples,:]

#g=sb.FacetGrid(data = Pisa_sample, col = 'GENDER',size=4)
for i in range(len(vars_target)):
    var=vars_target[i]
    g=sb.FacetGrid(data = Pisa_sample, col = 'GENDER',size=4)
    g.map(plt.scatter, var, 'ANCBELONG',alpha = 1/3)
    g.set_xlabels(Ab_dict[var].split("(")[0])
    g.set_ylabels("Sense of Belonging to School")
```
